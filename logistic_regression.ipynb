```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ticket_min_max = MinMaxScaler()
```

```python
ticket_min_max.fit(titan_data[['Fare']])
```

```python
min_max_data = ticket_min_max.transform(titan_data[['Fare']])
```

```python
min_max_df = pd.DataFrame(min_max_data)
```

```python
min_max_df
```

```python
min_max_df = min_max_df.rename({0: 'Fare MinMax Scaled'}, axis=1)
```

```python
titan_data = pd.concat([titan_data, min_max_df], axis = 1)
```

```python
titan_data
```

```python
X = titan_data[['Cabin Int', 'Gender Int','Pclass', 'Age', 'SibSp', 'Parch']]
y = titan_data['Survived']
```

```python
titan_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regression = LogisticRegression()
```

```python
logistic_regression.fit(X_train, y_train)
```

```python
predictions = logistic_regression.predict(X_test)
```

```python
predictions_df = pd.DataFrame({'Predictions': predictions, 'Actual': y_test})
```

```python
predictions_df
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
acc_score = accuracy_score(y_test, predictions)
```

```python
acc_score
```

```python
con_mat = confusion_matrix(y_test, predictions)
```

```python
con_mat
```

```python
class_repo = classification_report(y_test, predictions)
```

```python
print(class_repo)
```
